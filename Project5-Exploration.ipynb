# Data Wrangling

```python
#Create dataframe copy before wrangling
df = loans.copy()
```

<ul>
    <li>Get the average of Credit Score Range Upper and Lower and create a whole new column. Delete Credit Score Range Upper & Lower. </li>
    <li>Combine Other and Professional columns as 'Other Professionals'.</li>
    <li>Update Not Displayed Values to Not Reported.</li>
    <li>Delete rows with HR as values as that is a credit giving body and not a credit grade.</li>
<ul>

```python
#create a new column called CreditScores
df['CreditScore'] = (df.CreditScoreRangeUpper + df.CreditScoreRangeUpper) / 2

#Delete columns CreditScoreRangeUpper & CreditScoreRangeLower
df.drop(columns=['CreditScoreRangeUpper','CreditScoreRangeLower'], inplace=True)
```

```python
#Update Other and Professional Columns as 'Other Professionals'
mask = df.Occupation == 'Other'
column_name = 'Occupation'
df.loc[mask, column_name] = 'Other Professionals'

#Update Professional Columns as 'Other Professionals'
mask = df.Occupation == 'Professional'
column_name = 'Occupation'
df.loc[mask, column_name] = 'Other Professionals'
```

```python
#Update Not Displayed to Not reported
mask = df.IncomeRange == 'Not displayed'
column_name = 'IncomeRange'
df.loc[mask, column_name] = 'Not reported'
```

```python
#Drop HR values under Credit Grade
df = df[df.CreditGrade != 'HR']
```

### Univariate Data Exploration

#### Checking for Missing Values

```python
base_color = sns.color_palette()[9]
na_counts = df.isna().sum()
plt.figure(figsize=(10,20))
sns.barplot(na_counts, na_counts.index.values, color=base_color)
plt.xticks(rotation=90)
plt.ylabel('Loan Variables')
plt.xlabel('Count of Missing Values')
plt.title('Missing Values');
```

#### Exploration of Credit Scores

```python
#creating a countplot
base_color = sns.color_palette()[6]
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='CreditScore', color=base_color)
plt.title('Credit Score Count', fontsize=20);
```

> Credit Score has a normal distribution, indicating that the mean is higher in count compared to values further from the mean. The most common credit score is 699.

#### Exploration of Occupation Proportion

```python
#finding the proportion
n_loans = df.shape[0]
max_occupation = df.Occupation.value_counts()[0]
occupation_prop = max_occupation/n_loans

tick_props = np.arange(0, occupation_prop, 0.02)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]

#plotting
base_color = sns.color_palette()[3]
cat_order = df.Occupation.value_counts().iloc[0:10].index
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='Occupation', color=base_color, 
              order=cat_order, orient='v')

#creating ticks, labels and titles
plt.yticks(tick_props*n_loans, tick_names)
plt.xticks(label=tick_names, rotation=75)
plt.ylabel('Occupation Proportion')
plt.title('Top 10 Occupations', fontsize=20);
```

> The highest ratio of occupations in people with loans are indicated as Other Professionals, meaning that their professions are not specifically indicated in the list of occupations. This is followed by computer programmers, executives, teachers, administrative assistants, analysts, salespersons, accountants/CPAs and clerical jobs. 

> Due to the extremely high value of 'Other Professionals' and the fact that it provides no additional data, it should be dropped.

#### Exploration of Debt to Income Ratio

```python
#creating a log transform on a histogram
plt.figure(figsize=(12,6))
bin_edges = np.arange(0, df.DebtToIncomeRatio.max(), 0.1)
plt.hist(x=df.DebtToIncomeRatio, bins=bin_edges, rwidth=0.9)
plt.yscale('log')
tick_locs = [10,30,100,300,1000,3000,10000,30000]
plt.yticks(tick_locs, tick_locs)
plt.title('Debt to Income Ratio Counts', fontsize=20)
plt.xlim(0,4)
plt.ylabel('Count')
plt.xlabel('Ratio');
```

> According to the histogram, the ratio of debt to income of people with loans are right skewed, meaning that the mean is greater than the median. The predominant ratios are found between 0.1 to 0.5.

#### Exploration of Monthly Loan Payments

```python
plt.figure(figsize=(10,5))
plt.hist(df.MonthlyLoanPayment)
plt.xlim(0,1500)
plt.title('Monthly Loan Payments Count', fontsize=20)
plt.xlabel('Monthly Loan Payment')
plt.ylabel('Count');
```

> The distribution of monthly loan payments is right-skewed, with a predominant amount paying 200/month for loans.

#### Exploration of the Income Range

### Bivariate Data Exploration

#### Exploration of Relationship between Credit Score & Occupation

```python
#sorting only top 10 occupations, excluding 'Other Professionals'
occupations =df.Occupation.value_counts()[0:11]

#creating a list of occupations
occupation_list = []
for i in occupations.index:
    occupation_list.append(i)
    
#creating a dataframe that only has the top 20 under occupation
occupation_df = df[df['Occupation'].isin(occupation_list)]

#plotting
base_color = sns.color_palette()[4]
plt.figure(figsize=(12,16))
sns.violinplot(data=occupation_df, x='CreditScore', y='Occupation', color=base_color, inner='quartile')
plt.title('Relationship Between Occupations & Credit Scores', fontsize=20);
```

> Excluding 'Other Professionals', which made the plot highly skewed, the rest of the top 10 occupations have normal distributions. Most of the occupations hover between 650-750 in credit score. 

#### Visual relationship between Credit Score & Debt-to-Income Ratio

```python
#creating a sample size for credit scores
samples = np.random.choice(df.shape[0], 500, replace=False)
df_samp = df.loc[samples,:]

#plotting
plt.figure(figsize=(15,6))
base_color = sns.color_palette()[1]
sns.regplot(data=df_samp, x='CreditScore', y='DebtToIncomeRatio',color=base_color,  x_jitter=0.1, y_jitter=0.1)
plt.ylim(0,1)
plt.xlabel('CreditScore')
plt.ylabel('Debt To Income Ratio')
plt.title('Relationship between Credit Score & Debt To Income Ratio', fontsize=20);
```

> The scatter plot shows a very slightly negative linear relationship between debt-to-income ratio and credit scores with some obvious heteroskedestacity, which indicates the absensce of constant variance.

#### Visual Exploration of Credit Score & Monthly Loan Payment

```python
#Creating Violin Plots between Credit Score & Monthly Loan Payment

plt.figure(figsize=(10,6))
plt.hist2d(data=df, x='CreditScore', y='MonthlyLoanPayment', cmin=0.09,cmap='viridis_r')
plt.colorbar()
plt.xlabel('Credit Score')
plt.ylabel('Monthly Loan Payment')
plt.title('Relationship Between Credit Score & Monthly Loan Payment', fontsize=20);
```

> The heat map seems to show a positive relationship between credit scores and monthly loan payments. Credit scores between 700-725 with a monthly loan payment of less than 500 is the highest in occurence. It is also indicative of the fact that majority of bororowers pay 250-750/month.

#### Visual Relationship between Stated Monthly Income & Credit Score

```python
#creating function to transform data to plot regression line
#creating sample of 500 for clearer plotting
samples = np.random.choice(df.shape[0], 500, replace = False)
loans_samp = df.loc[samples,:]


plt.figure(figsize=(15,6))
sns.regplot(data=loans_samp, x='CreditScore', y='StatedMonthlyIncome',
            x_jitter=4, y_jitter=4,marker = 'x', color='g')
plt.ylim(0,10000)
plt.xlabel('CreditScore')
plt.ylabel('Stated Monthly Income')
plt.title('Relationship between Monthly Income & Credit Score', fontsize=20);
```

> According to the regression plot, albeit there is some obvious heteroskedestacity, Monthly Income and Credit Score have a positive relationship. I opted to limit the y-values up to 10,000 due to the severe outliers that make plotting less clear.

### Multivariate Data Exploration

#### Visual exploration of relationships between Occupations, Credit Scores & Debt to Income Ratio

```python
#divide occupations to top 5 and next 5 to avoid overplotting
occupation_list1 = occupation_list[0:6]
occupation_list2 = occupation_list[6:11]

#create custom dataframe that only contains top 5 occupations
occupation_top5 = df[df['Occupation'].isin(occupation_list1)]
occupation_next5 = df[df['Occupation'].isin(occupation_list2)]
```

```python
#plotting using occupation_df for top 5 occupations
g = sns.FacetGrid(data=occupation_top5, hue='Occupation', height=8, aspect=1.5, palette='icefire_r')
g = g.map(sns.regplot,'CreditScore','DebtToIncomeRatio', x_jitter=6, y_jitter=6, truncate=True)
g.add_legend()
plt.ylim([0,1])
plt.xlim([500,850])
plt.xlabel('Credit Scores')
plt.ylabel('Debt-To-Income Ratio');
```

```python
#plotting next 5 

g = sns.FacetGrid(data=occupation_next5, hue='Occupation', height=8, aspect=1.5, palette='icefire_r')
g = g.map(sns.regplot,'CreditScore','DebtToIncomeRatio', x_jitter=6, y_jitter=6, truncate=True)
g.add_legend()
plt.ylim([0,1])
plt.xlim([500,850])
plt.xlabel('Credit Scores')
plt.ylabel('Debt-To-Income Ratio');
```

> The regression plot is showing divided relationships between occupation, credit scores and debt to income ratio. In borrowers who work as administrative assistant, skilled laborer, salespeople(retail), retail management and clerks, it seems the relationship is positive whereas accountants/CPAs, executives, computer programmers, teachers, analysts and salespeople (commmission) trend towards negative.
